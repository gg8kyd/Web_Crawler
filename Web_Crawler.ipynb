```python
import requests
link = 'https://www.indeed.co.uk/jobs?q=Data+Scientist&l=London&radius=0&jt=fulltime'
website_raw = requests.get(link).text
```

```python
print(website_raw[0:1000])
print(len(website_raw))
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_raw, 'html.parser')
print(soup.prettify())
```

```python
title = soup.find(name = 'h2',attrs = {'class':'title'})
title
```

```python
my_data_list = soup.findAll('div',{'class':'jobsearch-SerpJobCard unifiedRow row result'})
title_list = []
company_list = []
summary_list = []
for ele in my_data_list:
    title_list.append(ele.find('h2',{'class':'title'}).find('a').get('title'))
    company_list.append(ele.find('span',{'class':'company'}).get_text().strip())
    summary_list.append(ele.find('div',{'class':'summary'}).get_text().strip())
```

```python
import pandas as pd
df = pd.DataFrame()
df['title'] = title_list
df['company'] = company_list
df['summary'] = summary_list
```

```python
df
```

```python
df.to_csv('test10.csv')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=None,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))
    
    fig = plt.figure(1, figsize=(8, 8))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

```python
show_wordcloud(df.summary, title = "job summary")
```
